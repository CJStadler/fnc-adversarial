```python
import matplotlib.pyplot as plt
```

```python
def plot_model(points, **kwargs):
    x, y = zip(*points)
    plt.plot(x, y, **kwargs)
```

```python
# TODO: fill in actual values
# Lists of points (N, accuracy %)
baseline = [(0, 80), (3, 70), (10, 50)]
uclmr = [(0, 89), (3, 85), (10, 80)]

plot_model(baseline, label='Baseline')
plot_model(uclmr, label='UCL MR')

plt.ylabel('Accuracy')
plt.xlabel('Number of Changes')
plt.legend()
plt.show()
```
